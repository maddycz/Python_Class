```python
#drops rows that do not have info on at least 13 columns
df.dropna(thresh=13)
df_drop=df.dropna(thresh=13)
df_drop
```

```python
#matching for mean score in order to remove the blank 's'
df[df['Mean Score'] == 's']
df_MeanScore=df[df['Mean Score'] == 's']
df_MeanScore
```

```python
#removing the rows with 's' as mean score
df_new=df.drop(df_MeanScore.index)
df_new
```

```python
#limiting Search to Common Core Algebra
df_new[df_new['Regents Exam']== 'Common Core Algebra']
df_Algebra=df_new[df_new['Regents Exam']== 'Common Core Algebra']
df_Algebra.head(20)
```

```python
#filtering by year 2015 only
df_Algebra[df_Algebra['Year']==2015]
df_2015=df_Algebra[df_Algebra['Year']==2015]
df_2015
```

```python
#limiting to K-8 schools only
df_2015[df_2015['School Level']== 'K-8']
df_K8=df_2015[df_2015['School Level']== 'K-8']
df_K8.head (20)
```

```python
#subset of School 01m184 only
df_K8[df_K8 ['School DBN']=='01M184']
df_K8s1=df_K8[df_K8 ['School DBN']=='01M184']
df_K8s1
```

```python
#bar chart of mean score for school P.S. 184m for 2015 Common Core Algebra
df_K8s1.sort_values('Mean Score').plot(kind="bar", y= 'Total Tested', x= 'Mean Score', color = 'orange')
plt.title("P.S. 184m Shuang Wen 2015 Common Core Algebra Regents Results")
plt.xlabel("Mean Score")
plt.ylabel("Total Tested")
```

```python
#pie chart of Percent Scoring Below 66 at P.S. 184m for 2015 Common Core Algebra
df_K8s1.groupby(['Percent Scoring Below 65']).sum().plot(kind='pie', y='Total Tested', autopct='%1.0f%%')
plt.title('Percent of Students Scoring Below 65 in P.S. 184m')
```

```python
#pie chart of Percent Scoring 65 or Above at P.S. 184m for 2015 Common Core Algebra
df_K8s1.groupby(['Percent Scoring 65 or Above']).sum().plot(kind='pie', y='Total Tested', autopct='%1.0f%%')
plt.title('Percent of Students Scoring 65 or Above in P.S. 184m')
```

```python
#pie chart of Percent Scoring 80 or Above at P.S. 184m for 2015 Common Core Algebra
df_K8s1.groupby(['Percent Scoring 80 or Above']).sum().plot(kind='pie', y='Total Tested', autopct='%1.0f%%')
plt.title('Percent of Students Scoring 80 or Above in P.S. 184m')
```
