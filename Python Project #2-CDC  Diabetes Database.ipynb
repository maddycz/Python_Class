```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#removing the '?'
missing_values=['?']
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv',na_values=missing_values)
```

```python
#displaying large protion of rows and columns
pd.options.display.max_rows = 500
df.head(500)
```

```python
#replacing 'NaN' in race column to 'Other'
df['race'].fillna('Other', inplace=True)
df.head(30)
```

```python
#limiting Search to Age 50-60
df[df['age']== '[50-60)']
df_Age=df[df['age']== '[50-60)']
df_Age.head(30)
```

```python
df_Age.shape
```

```python
#grouping by gender and race to get demographic counts
df_Age.groupby (['race'])['gender'].value_counts()
```

```python
df_Age2=df_Age[['race', 'gender', 'age']]
df_Age2
```

```python
df_Age2.info()
```

```python
#subsetting to African Americans Only
df_AA=df_Age2[df_Age2['race'] == 'AfricanAmerican']
df_AA
```

```python
#distinguishing count of African American Females and Males Ages 50-60 with Diabetes
df_AA.groupby (['gender']).count()
```

```python
#getting percentage of Afican Amercian Males vs Female within 50-60 age range that have Diabetes
df_AA['gender'].value_counts(normalize=True)*100
```

```python
#Displaying pie chart of proportion of African American Female/Males that have Diabetes within the 50-60 Age Range
df_AA['gender'].value_counts(normalize=True).plot(kind='pie', autopct='%1.0f%%')
plt.title('Percentage of African American Female/Male 50-60 Years Old with Diabetes')
```
